### Analyze the columns which indicate the dissatisfaction for the job within the institute. 

```python
dete_resignations.info()
```

<font color = 'dark orange'>__DETE SURVEY DATASET has the following columns indicating the Job dissatisfactions.
Filter the data on these columns and check for observations which have atleast one column with dissatisfaction indicated __</font>

```python
dissatisfaction_columns = ["job_dissatisfaction", 
                           "dissatisfaction_with_the_department",
                           "physical_work_environment", 
                           "lack_of_recognition",
                           "lack_of_job_security",
                           "work_location",
                           "employment_conditions",
                           "work_life_balance",
                           "workload"]                          

dete_resignations["dissatisfied"] = dete_resignations[dissatisfaction_columns].any(axis='columns', skipna=False)
print(dete_resignations.shape)
dete_resignations["dissatisfied"].value_counts(dropna=False)
```

<font color = 'dark orange'>__TAFE SURVEY DATSET has the following columns indicating the Job dissatifactions. Filter the data on these columns and check for observations which have atleast one column with dissatisfaction indicated__</font>

```python
dissatisfaction_columns = ['Contributing Factors. Dissatisfaction',
                           'Contributing Factors. Job Dissatisfaction']

tafe_resignations[dissatisfaction_columns]
```

1. Need to transform the dissatisfaction column values to either True, False or NaN
  1. Change the NaN value to np.nan value
  2. Change the '-' value to False
  3. Any Other value to be True

```python
# Function to transform the values of the column
def col_transform(value):
    if pd.isnull(value):
        return np.NaN
    elif value == "-":
        return False
    else:
        return True
    
tafe_resignations[dissatisfaction_columns] = tafe_resignations[dissatisfaction_columns].applymap(col_transform)

tafe_resignations[dissatisfaction_columns]
```

<font color='dark orange'>__TAFE SURVEY DATSET Filter the data on these columns and check for observations which have atleast one column with dissatisfaction indicated__</font>

```python
# Create a dissatisfied column to find if any of the columns contributed to the Job dissatisfactions
tafe_resignations["dissatisfied"] = tafe_resignations[dissatisfaction_columns].any(axis='columns', skipna=False)
tafe_resignations["dissatisfied"].value_counts(dropna=False)
```

```python
# Make a copy of the resignations dataframe
dete_resignations_up = dete_resignations.copy()
tafe_resignations_up = tafe_resignations.copy()
```

#### <font color='dark orange'>Add the institute name to the respective survey dataset to identify which dataset the date belongs to</font>

```python
dete_resignations_up["institute"] = "DETE"
tafe_resignations_up["institute"] = "TAFE"
```

<font color="dark orange" __COMBINE THE DETE AND TAFE DATASETS__  </font>

```python
#print(dete_resignations_up.shape)
#print(tafe_resignations_up.shape)

combined = pd.concat([dete_resignations_up, tafe_resignations_up], axis='rows', join='outer')

# Shape of the dataframe
print(combined.shape)

# Display the rows of the combined dataframe
combined.head(3)
```

__<font color='dark orange'>DROP COLUMNS FROM DATASET WHICH CONTAIN LESS THAN 500 NON_NULL VALUES__</font>

```python
combined_updated = combined.dropna(axis=1, thresh=500).copy()
combined_updated
```

#### The DETE and TAFE DATASETS ARE COMBINED, THE INSTITUTE SERVCIES FEATURE CAN BE USED TO AGGREGATE THE NUMBER OF DISSATISFIED EMPLOYEES GROUPED WITH THE SURVEY TYPE AS WELL 

```python
combined['institute_service'].value_counts(dropna=False)
```

<font color='dark orange'>__Calculate the number of years of service for the employee__</font>

```python
# Extract the number of years from the 'institute service' column: Define a function to calculate the field
pattern = r"(\d+)"
combined_updated["institute_service_up"] = combined_updated["institute_service"].astype(str).str.extract(pattern, expand=False)
combined_updated["institute_service_up"] = combined_updated["institute_service_up"].astype(float)
combined_updated["institute_service_up"].value_counts()
```

__<font color='dark orange'>Identify which category (experience level bandwidth) does the Employee belongs to</font>__

```python
def service_category(val):
    if pd.isnull(val):
        return np.nan
    elif val < 3:
        return "New"
    elif val >= 3 and val < 7:
        return "Experienced"
    elif val >= 7 and val < 11:
        return "Established"
    elif val >= 11:
        return "Veteran"

combined_updated["service_cat"] = combined_updated["institute_service_up"].apply(service_category)
combined_updated["service_cat"].value_counts(dropna = False)
```

```python
# Check for the possible values in the 'dissastisfied' series
combined_updated["dissatisfied"].value_counts(dropna=False)
```

__<font color='dark orange'>Fill in the missing values for this column and the most easy way to fill is to be replaced by the max occurance of the valid values (which is True)</font>__

```python
combined_updated["dissatisfied"].fillna(False, inplace=True)
```

```python
# Check the previous operation
combined_updated["dissatisfied"].value_counts(dropna=False)
```

__Pivot table to view the percentage of dissatisfied employees based on the service category (lower duration vs longer duration)__

```python
dissatisfaction_pct = combined_updated.pivot_table(values="dissatisfied", index="service_cat")
dissatisfaction_pct
#type(dissatisfaction_pct)
axis = ['Established', 'Experienced', 'New', 'Veteran']
```

```python
# Let's publish the results on a graph
#dissatisfaction_pct.plot.bar(rot=30, color=('red', 'beige', 'beige', 'beige'), axis='off')
#### The results show that Established 
```

```python
y_pos = np.arange(dissatisfaction_pct['dissatisfied'].size)
plt.bar(y_pos-0.25, dissatisfaction_pct['dissatisfied']*100, width=0.3, color=['red', 'ivory', 'ivory', 'ivory'])
plt.grid(False)
plt.xticks(y_pos-0.25, list(dissatisfaction_pct.index), rotation=30)
plt.xlabel('Service category', fontweight='bold', color = 'orange', fontsize='18')
plt.ylabel('Dissatisfaction %', fontweight='bold', color = 'red', fontsize='18')
plt.style.use('seaborn-dark')
plt.show()
```
